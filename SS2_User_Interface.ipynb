```python
import random
import tkinter as tk
from tkinter import messagebox, ttk
from PIL import Image, ImageTk

# Initializes a Tkinter application window titled 'African Quiz' for creating a graphical user interface (GUI) for the quiz
root = tk.Tk()
root.title("African Quiz")
root.config(bg='#dadada')

# Custom theme and design elements
font_title = ("Arial", 24, "bold")
font_question = ("Arial", 18)
font_options = ("Arial", 18)

# Load image
image_path = "African Continent.jfif"  
image = Image.open(image_path)
image.thumbnail((300, 200))  
photo = ImageTk.PhotoImage(image)

# Display image
image_label = tk.Label(root, image=photo, bg='#dadada')
image_label.image = photo  
image_label.pack(anchor='w', padx=10, pady=10)

# Welcome message and unique username
names = ['HeritageHunter', 'MandelaMind', 'NileNavigator', 'BaobabBrain', 'MaliMystic']
username = random.choice(names) + str(random.randint(100, 999))
welcome_text = f"Welcome to the African Trivia Quiz!\n\nHello, {username}!"

# Create welcome label
welcome_label = tk.Label(root, text=welcome_text, font=font_title, bg='#dadada', fg='#2C6E63', justify="left", wraplength=800)
welcome_label.pack(anchor='w', padx=10, pady=10)

# Hide welcome message and image after some seconds
def hide_welcome():
    welcome_label.pack_forget()
    image_label.pack_forget()

root.after(8000, hide_welcome)

#Quiz Questions Data Structure
quiz_questions = [
    {
        'category': 'History',
        'difficulty_level' : 'Beginner',
        'question' : 'What year did South Africa gain independence from Britain?',
        'options' : ['A. 1910', 'B. 1945', 'C. 1961', 'D. 1994'],
        'correct_answer' : 'C'
    },
    {
        'category': 'History',
        'difficulty_level': 'Beginner',
        'question': 'What was the main export of the Mali Empire?',
        'options': ['A. Gold', 'B. Salt', 'C. Ivory', 'D. Diamonds'],
        'correct_answer': 'A'
    },
    {
        'category': 'History',
        'difficulty_level': 'Beginner',
        'question': 'The Great Zimbabwe was a city in the historical African kingdom of __________.',
        'options': ['A. The Kingdom of Mapungubwe', 'B. The Kingdom of Zimbabwe', 'C. The Kingdom of Mutapa', 'D. The Kingdom of Butua'],
        'correct_answer': 'B'
    },
    {
        'category': 'History',
        'difficulty_level': 'Moderate',
        'question': 'Kwame Nkrumah was the first president of Ghana.',
        'options': ['A. True', 'B. False'],
        'correct_answer': 'A'
    },
    {
        'category': 'History',
        'difficulty_level': 'Moderate',
        'question': 'The Battle of Adwa was a victory for which African country against Italy in 1896?',
        'options': ['A. Ethiopia', 'B. Eritrea', 'C. Somalia', 'D. Sudan'],
        'correct_answer': 'A'
    },
    {
        'category': 'History',
        'difficulty_level': 'Moderate',
        'question': 'The Ashanti Empire was located in which modern-day country?',
        'options': ['A. Nigeria', 'B. Ghana', 'C. Ivory Coast', 'D. Burkina Faso'],
        'correct_answer': 'B'
    },
    {
        'category': 'History',
        'difficulty_level': 'Advanced',
        'question': 'The Kingdom of Kush was located in modern-day __________.',
        'options': ['A. Egypt', 'B. Sudan', 'C. Ethiopia', 'D. Libya'],
        'correct_answer': 'B'
    },
    {
        'category': 'History',
        'difficulty_level': 'Advanced',
        'question': 'Cetshwayo kaMpande led the Zulu Kingdom during the Anglo-Zulu War.',
        'options': ['A. True', 'B. False'],
        'correct_answer': 'A'
    },
    {
        'category': 'History',
        'difficulty_level': 'Advanced',
        'question': 'The Axum Empire, one of the four greatest civilizations in the world, was located in which modern-day country?',
        'options': ['A. Egypt', 'B. Ethiopia', 'C. Sudan', 'D. South Africa'],
        'correct_answer': 'B'
    },
    {
        'category': 'History',
        'difficulty_level': 'Challenging',
        'question': 'Arrange the following empires in order of their historical emergence, from earliest to latest: Mali Empire, Kingdom of Kush, Songhai Empire, and Kingdom of Benin',
        'options': [
            'A. Mali Empire, Kingdom of Kush, Kingdom of Benin, Songhai Empire',
            'B. Kingdom of Kush, Kingdom of Benin, Mali Empire, Songhai Empire',
            'C Kingdom of Benin, Kingdom of Kush, Songhai Empire, Mali Empire',
            'D Songhai Empire, Kingdom of Benin, Mali Empire, Kingdom of Kush'
        ],
        'correct_answer': 'B'
    },
    {
        'category': 'History',
        'difficulty_level': 'Challenging',
        'question': 'Identify the correct statement regarding the impact of colonialism on African societies:',
        'options': [
            'A. Colonialism led to the preservation of indigenous cultures and traditions.',
            'B. African societies experienced rapid industrialization under colonial rule.',
            'C. Colonial borders often divided ethnic groups, leading to conflicts that persist to this day.',
            'D. Colonial powers provided equal rights and opportunities to all Africans.'
        ],
        'correct_answer': 'C'
    },
    {
        'category': 'History',
        'difficulty_level': 'Challenging',
        'question': 'Fill in the blank with the correct term: During the ________________, European powers met to carve up Africa, leading to the establishment of colonial borders that often disregarded indigenous territories and ethnic groups.',
        'options': ['A. Slave Trade', 'B. Scramble for Africa', 'C. Great Migration', 'D. Pan-African Congress'],
        'correct_answer': 'B'
    },

    {
        'category': 'Culture',
        'difficulty_level': 'Beginner',
        'question': 'What is the traditional religion of the Yoruba people?',
        'options': ['A. Christianity', 'B. Islam', 'C. Yoruba religion', 'D. Hinduism'],
        'correct_answer': 'C'
    },
    {
        'category': 'Culture',
        'difficulty_level': 'Beginner',
        'question': 'The Maasai people are native to Kenya and Uganda.',
        'options': ['A. True', 'B. False'],
        'correct_answer': 'B'
    },
    {
        'category': 'Culture',
        'difficulty_level': 'Beginner',
        'question': 'What is the traditional method of communication among the San people of Southern Africa?',
        'options': ['A. Drumming', 'B. Smoke signals', 'C. Petroglyphs', 'D. Oral storytelling'],
        'correct_answer': 'D'
    },
    {
        'category': 'Culture',
        'difficulty_level': 'Moderate',
        'question': 'The Dogon people, known for their astronomical knowledge, are from which country?',
        'options': ['A. Mali', 'B. Burkina Faso', 'C. Niger', 'D. Chad'],
        'correct_answer': 'A'
    },
    {
        'category': 'Culture',
        'difficulty_level': 'Moderate',
        'question': 'The Adinkra symbols originate from which African culture?',
        'options': ['A. Ashanti', 'B. Zulu', 'C. Maasai', 'D. Yoruba'],
        'correct_answer': 'A'
    },
    {
        'category': 'Culture',
        'difficulty_level': 'Moderate',
        'question': 'The Kente cloth is native to which African country?',
        'options': ['A. Nigeria', 'B. Ghana', 'C. Senegal', 'D. Ivory Coast'],
        'correct_answer': 'B'
    },
    {
        'category': 'Culture',
        'difficulty_level': 'Advanced',
        'question': 'The Griot tradition of storytelling is most associated with East Africa.',
        'options': ['A. True', 'B. False'],
        'correct_answer': 'B'
    },
    {
        'category': 'Culture',
        'difficulty_level': 'Advanced',
        'question': 'What is the name of the traditional Ethiopian communal serving dish often used in traditional meals?',
        'options': ['A. Tagine', 'B. Injera', 'C. Jollof', 'D. Ugali'],
        'correct_answer': 'B'
    },
    {
        'category': 'Culture',
        'difficulty_level': 'Advanced',
        'question': 'The Sankofa symbol, which represents the importance of learning from the past, originates from which African culture?',
        'options': ['A. Zulu', 'B. Maasai', 'C. Ashanti', 'D. Yoruba'],
        'correct_answer': 'C'
    },
    {
        'category': 'Notable Figures',
        'difficulty_level': 'Beginner',
        'question': 'Who is the South African anti-apartheid revolutionary who served as President of South Africa from 1994 to 1999?',
        'options': ['A. Desmond Tutu', 'B. F. W. de Klerk', 'C. Nelson Mandela', 'D. Thabo Mbeki'],
        'correct_answer': 'C'
    },
    {
        'category': 'Notable Figures',
        'difficulty_level': 'Beginner',
        'question': 'Ellen Johnson Sirleaf was the first female president in Africa.',
        'options': ['A. True', 'B. False'],
        'correct_answer': 'A'
    },
    {
        'category': 'Notable Figures',
        'difficulty_level': 'Beginner',
        'question': 'Christiaan Barnard performed the world’s first successful heart transplant.',
        'options': ['A. True', 'B. False'],
        'correct_answer': 'A'
    },
    {
        'category': 'Notable Figures',
        'difficulty_level': 'Moderate',
        'question': 'Who was the first black Nobel laureate?',
        'options': ['A. Nelson Mandela', 'B. Martin Luther King Jr.', 'C. Albert Luthuli', 'D. Desmond Tutu'],
        'correct_answer': 'C'
    },
    {
        'category': 'Notable Figures',
        'difficulty_level': 'Moderate',
        'question': 'Chinua Achebe is the author of “__________”.',
        'options': ['A. Things Fall Apart', 'B. The Lion and the Jewel', 'C. Anthills of the Savannah', 'D. Arrow of God'],
        'correct_answer': 'A'
    },
    {
        'category': 'Notable Figures',
        'difficulty_level': 'Moderate',
        'question': 'Who is the Kenyan woman who won the Nobel Peace Prize in 2004?',
        'options': ['A. Wangari Maathai', 'B. Ellen Johnson Sirleaf', 'C. Joyce Banda', 'D. Ameenah Gurib'],
        'correct_answer': 'A'
    },
    {
        'category': 'Notable Figures',
        'difficulty_level': 'Advanced',
        'question': 'Who was the Ghanaian independence leader and the first Prime Minister and President of Ghana?',
        'options': ['A. Patrice Lumumba', 'B. Kwame Nkrumah', 'C. Jomo Kenyatta', 'D. Julius Nyerere'],
        'correct_answer': 'B'
    },
    {
        'category': 'Notable Figures',
        'difficulty_level': 'Advanced',
        'question': 'Desmond Tutu is the South African social rights activist and retired Anglican bishop who won the Nobel Peace Prize in __________.',
        'options': ['A. 1980', 'B. 1984', 'C. 1990', 'D. 1994'],
        'correct_answer': 'B'
    },
    {
        'category': 'Notable Figures',
        'difficulty_level': 'Advanced',
        'question': 'Ken Saro-Wiwa was the Nigerian environmental activist who led the Movement for the Survival of the Ogoni People (MOSOP).',
        'options': ['A. True', 'B. False'],
        'correct_answer': 'A'
    },
]

# Implement Functions
def generate_question(category, difficulty):
    """
    Generate a random question based on the specified category and difficulty level.
    """
    filtered_questions = []
    for q in quiz_questions:
        if q['category'] == category and q['difficulty_level'] == difficulty:
            filtered_questions.append(q)
    return random.choice(filtered_questions)

def validate_answer(question, player_answer):
    """
    Check if the player's answer is correct by comparing it with the 'correct_answer' key in the question dictionary.
    """
    return player_answer == question['correct_answer']

def update_score(score, is_correct):
    """
    Update the player's score based on the correctness of the answer.
    """
    if is_correct:
        return score + 1
    else:
        return score


def display_question_and_options(question, question_label, options_label, submit_button, answer_var):
    """
    Display the question and options along with a submit button.
    """
    question_label.config(text=question['question'])
    for widget in options_label.winfo_children():
        widget.destroy()  
    for i, option in enumerate(question['options']):
        rb = tk.Radiobutton(options_label, text=option, variable=answer_var, value=chr(65 + i),
                            font=font_options, bg='#dadada', fg='#024D7A', selectcolor='#dadada',
                            activebackground='#dadada', activeforeground='#dadada')
        rb.pack(anchor='w')
    submit_button.config(command=lambda: submit_answer(question, answer_var))

def show_final_score():
    final_score = int(score_label.cget('text').split(':')[1].split('/')[0])
    feedback = ""
    # Evaluate final score and provide feedback
    if final_score <= 3:
        feedback = f"{username}, you have very basic knowledge. Keep learning and improving!"
    elif final_score <= 6:
        feedback = f"{username}, you have a moderate level of knowledge. Good job!"
    elif final_score <= 8:
        feedback = f"{username}, you have a good understanding of the material. Well done!"
    elif final_score <= 9:
        feedback = f"{username}, you have an excellent understanding of the material. Great work!"
    else:
        feedback = f"Perfect score, {username}! You have outstanding knowledge!"
    
    messagebox.showinfo("Quiz completed!", f"Quiz completed!\n\nFinal Score: {final_score}/{total_levels}\n\n{feedback}")

def submit_answer(question, answer_var):
    """
    Validate the player's answer, update score, and display message before proceeding to the next question.
    """
    player_answer = answer_var.get()
    is_correct = validate_answer(question, player_answer)
    score = int(score_label.cget("text").split(":")[1].split("/")[0])
    score_label.config(text=f"Score: {update_score(score, is_correct)}/{total_levels}")
    
    # Display congratulatory or incorrect message based on correctness of answer
    if is_correct:
        congratulatory_messages = [
            f"Bravo, {username}! You've nailed it!",
            f"Outstanding, {username}! Your answer is spot on!",
            f"You're on fire, {username}! That's the right answer!",
            f"Fantastic job, {username}! You've chosen wisely!",
            f"Incredible, {username}! You've cracked it!"
        ]
        messagebox.showinfo("Correct!", random.choice(congratulatory_messages))
    else:
        incorrect_messages = [
            f"Oops, {username}! That's not quite right. The correct answer is:",
            f"Better luck next time, {username}! The correct answer is:",
            f"Incorrect, {username}. The correct answer is:",
            f"Not quite there, {username}. The correct answer is:",
            f"Almost, {username}! The correct answer is:"
        ]
        incorrect_message = random.choice(incorrect_messages)
        correct_option = question['correct_answer']
        messagebox.showinfo("Incorrect!", f"{incorrect_message} {correct_option}")

    # Proceed to the next question or end the quiz
    global level
    level += 1
    if level <= total_levels:
        next_question = set_level()
        level_label.config(text=f"Level {level}")  
        display_question_and_options(next_question, question_label, options_label, submit_button, answer_var)
    else:
        show_final_score()

    # Update progress bar based on current level
    progress_bar["value"] = (level / total_levels) * 100

    
# Define levels
def set_level():
    if level == 1:
        return generate_question('History', 'Beginner')
    elif level == 2:
        return generate_question('Culture', 'Beginner')
    elif level == 3:
        return generate_question('Notable Figures', 'Beginner')
    elif level == 4:
        return generate_question('History', 'Moderate')
    elif level == 5:
        return generate_question('Culture', 'Moderate')
    elif level == 6:
        return generate_question('Notable Figures', 'Moderate')
    elif level == 7:
        return generate_question('History', 'Advanced')
    elif level == 8:
        return generate_question('Culture', 'Advanced')
    elif level == 9:
        return generate_question('Notable Figures', 'Advanced')
    elif level == 10:
        return generate_question('History', 'Challenging')
    else:
        return None

# Load the image for the banner
banner_image_path = "Banner.png"  
banner_image = Image.open(banner_image_path)
banner_image.thumbnail((800, 300))  
banner_photo = ImageTk.PhotoImage(banner_image)

# Create Tkinter widgets
level_label = tk.Label(root, text="Level 1", font=font_title, bg='#dadada', fg='#010101')
username_label = tk.Label(root, text=f"Username: {username}", font=font_options, bg='#dadada', fg='#010101')
question_label = tk.Label(root, text="", font=font_question, bg='#dadada', fg='#1D6085', wraplength=900)
options_label = tk.Label(root, bg='#dadada')
score_label = tk.Label(root, text="Score: 0/1", font=font_options, bg='#dadada', fg='#EA5B07')
answer_var = tk.StringVar(root)  
submit_button = tk.Button(root, text="Submit", font=font_options, bg='#1D6085', fg='#dadada')
progress_bar = ttk.Progressbar(root, orient="horizontal", length=200, mode="determinate")
progress_label = tk.Label(root, text="Progress Bar", bg="#dadada", fg="#F3A10D", font='14')        
banner_label = tk.Label(root, image=banner_photo, bg='#dadada')
banner_label.image = banner_photo  

# Layout Tkinter widgets
welcome_label.pack(anchor='w', padx=10, pady=10)
level_label.pack(anchor='w', padx=10, pady=10)
username_label.pack(anchor='w', padx=10, pady=10)
question_label.pack(anchor='w', padx=10, pady=10)
options_label.pack(anchor='w', padx=10, pady=10)
score_label.pack(anchor='e', padx=10, pady=10)
submit_button.pack(anchor='w', pady=5)
progress_bar.pack(anchor='w', padx=10, pady=10)
progress_label.pack(anchor='w', padx=10, pady=(0, 10))
banner_label.pack(anchor='center', pady=10)

# Set total_levels to the desired number of quiz levels
total_levels = 10
level = 1  

# Display the first question
first_question = set_level()
display_question_and_options(first_question, question_label, options_label, submit_button, answer_var)

# Run the Tkinter event loop
root.mainloop()
```
